```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 데이터 로드 (예시: CSV 파일)
# data = pd.read_csv("your_data.csv")
data = df.copy()

# 특징 선택 (필요한 열만 선택)
# features = data[['feature1', 'feature2', ...]]
features = df.iloc[:, :1]
# features = df.iloc[:,1]
# 데이터 표준화
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# K-Means 클러스터링 (K 값 설정)
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(scaled_features)

# 각 데이터 포인트가 속한 클러스터 할당
labels = kmeans.labels_

# 클러스터 중심 계산
centers = kmeans.cluster_centers_

# 각 데이터 포인트와 해당 클러스터 중심 사이의 거리 계산
distances = kmeans.transform(scaled_features)

# 임계값 설정 (예시: 평균 거리의 2배)
threshold = np.mean(distances) * 2

# 임계값을 초과하는 데이터를 이상치로 판별
outliers = np.where(distances.min(axis=1) > threshold)[0]

# 이상치 데이터 출력
print("Outliers:", data.iloc[outliers])
```

1. K-Means 클러스터링 분석 결과

34만 개의 데이터 중 약 2.2% (7700개)를 이상치로 파악했다는 결과는, K-Means 클러스터링을 사용해 데이터를 군집화한 후 특정 군집(클러스터)을 이상치로 간주했음을 의미합니다.
이상치는 주로 클러스터 중심에서 멀리 떨어져 있는 데이터 포인트로 정의되었을 가능성이 큽니다.
이 과정은 **이상치 탐지(Anomaly Detection)**의 첫 번째 단계로 사용되며, 주어진 데이터에서 "비정상적"으로 간주되는 패턴을 가진 데이터를 걸러내는 데 유용합니다.
2. 이상치를 제외한 데이터로 Autoencoder 학습

K-Means로 이상치를 제거하고 나머지 데이터(정상 데이터)를 Autoencoder 모델로 학습시켰습니다.
Autoencoder는 입력 데이터를 재구성(reconstruct)하는 딥러닝 모델로, 정상 데이터의 특징을 잘 학습합니다.
이상치 제외 데이터로 학습했기 때문에, 모델은 정상 데이터만 잘 재구성할 수 있도록 최적화됩니다.
3. Reconstruction Threshold (재구성 임계값): 0.7081

Autoencoder 모델은 학습한 정상 데이터를 재구성할 때 재구성 오차(Reconstruction Error)가 낮습니다.
재구성 오차는 일반적으로 입력 데이터와 모델 출력(재구성 결과) 간 차이의 절댓값이나 제곱합으로 계산됩니다.
정상 데이터로 학습했기 때문에, 정상 데이터는 재구성 오차가 작고, 이상치는 모델이 잘 재구성하지 못하므로 오차가 크게 나타납니다.
Reconstruction Threshold: 0.7081은 정상 데이터와 이상치를 구분하는 기준값을 의미합니다.
재구성 오차가 0.7081 이하인 데이터는 정상으로 간주됩니다.
재구성 오차가 0.7081 초과인 데이터는 이상치로 간주됩니다.

Autoencoder 학습과 Reconstruction Threshold:

Autoencoder 모델이 정상 데이터를 학습한 후, 재구성 오차를 기준으로 정상과 이상을 구분합니다.
Threshold인 0.7081은 정상 데이터와 이상치를 구분하는 경계값입니다.
이 과정의 목적:

K-Means와 Autoencoder의 결합은 이상치 탐지의 정밀도를 높이기 위한 과정입니다.
K-Means로 초기 이상치 제거를 하고, Autoencoder로 정교한 이상치 탐지를 수행하여 이상치 탐지의 신뢰성을 높였습니다.
실용적 해석:

정상 데이터로 학습한 Autoencoder는 새로운 데이터에서 재구성 오차를 계산하여, 재구성 오차가 0.7081 이상인 데이터는 이상치로 판단합니다.

```python
import numpy as np
import pandas as pd
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
)
import matplotlib.pyplot as plt
# 정상 테스트 데이터에 대한 재구성 오차 계산
x_test_normal_pred = AE_model.predict(X_normal_test)
test_normal_mae_loss = np.mean(np.abs(x_test_normal_pred - X_normal_test), axis=1)


# 비정상 테스트 데이터에 대한 재구성 오차 계산
x_test_abnormal_pred = AE_model.predict(X_abnormal_test)
test_abnormal_mae_loss = np.mean(np.abs(x_test_abnormal_pred - X_abnormal_test), axis=1)

# 정상/비정상 데이터 라벨링
y_test_normal = np.zeros(len(test_normal_mae_loss))  # 정상 데이터는 0으로 라벨링
y_test_abnormal = np.ones(len(test_abnormal_mae_loss))  # 비정상 데이터는 1로 라벨링

# 테스트 데이터 재구성 오차 합치기
y_true = np.concatenate([y_test_normal, y_test_abnormal])  # 실제 라벨 (0 = 정상, 1 = 비정상)
y_pred = np.concatenate([
    (test_normal_mae_loss > threshold).astype(int),  # 정상 데이터에 대한 예측 (Threshold 초과 시 1)
    (test_abnormal_mae_loss > threshold).astype(int)  # 비정상 데이터에 대한 예측
])

# Precision, Recall, F1 Score 계산
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

# 결과 출력
print("Precision (정밀도):", precision)
print("Recall (재현율):", recall)
print("F1 Score:", f1)

# 추가적으로 Classification Report 출력
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=["정상", "비정상"]))
# 혼동행렬 계산
conf_matrix = confusion_matrix(y_true, y_pred)
print(conf_matrix)
```
